```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = df.copy()

# Bar chart for 'Survived' count
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=titanic_data, palette='viridis', hue='Survived', legend=True)
#plt.legend(title='Survived', loc='upper right')
plt.title('Count of Survivors and Non-Survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for 'Survived' count by "Sex"
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=titanic_data, palette='viridis')
plt.title('Count of Survivors and Non-Survivors by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for 'Survived' count by "Pclass"
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_data, palette='viridis')
plt.title('Count of Survivors and Non-Survivors by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```
